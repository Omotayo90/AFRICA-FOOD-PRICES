## Table query

For the western and eastern part of of Africa

```python
pivot_table = pd.pivot_table(df, index=['country', 'year', 'month', 'market'], values = ['price'], aggfunc='mean')
```

```python
eastern_region = pivot_table.query("country == ['Uganda', 'Kenya']")
western_region = pivot_table.query("country == ['Benin', 'Cape Verde', 'Gambia', 'Ghana', 'Guinea', 'Guinea-Bissau', 'Liberia', 'Niger', 'Nigeria', 'Senegal', 'Sierra Leone', 'Togo']")                                   
```

```python
easternRegionResetIndex = eastern_region.reset_index()
westernRegionResetIndex = western_region.reset_index()
```

## Prices Of Commodities In Eastern Region

```python
eastXaxis = easternRegionResetIndex.country
eastYaxis = easternRegionResetIndex.price

plt.bar(eastXaxis,eastYaxis)
plt.title("The Minimum Prices Of Commodities In Eastern Region")
plt.xlabel("Eastern Countries")
plt.ylabel("Commodities Minimum Prices")
plt.xticks(rotation=50)
plt.show()
```

```python
eastdataset = easternRegionResetIndex
plt.figure(figsize =(12,6))
plt.title("Prices Trends Over The Years")
sns.lineplot(x="year",y="price",data=eastdataset)
plt.xlabel("Years")
plt.ylabel("Prices")
plt.show()
```

## Prices Of Commodities In Western Region

```python
westXaxis = westernRegionResetIndex.country
westYaxis = westernRegionResetIndex.price

plt.bar(westXaxis,westYaxis)
plt.title("The Minimum Prices Of Commodities In Western Region")
plt.xlabel("Western Countries")
plt.ylabel("Minimum Prices of Commodities")
plt.xticks(rotation=50)
plt.show()
```

```python
westdataset = westernRegionResetIndex
plt.figure(figsize =(12,6))
plt.title("Prices Trends Over The Years")
sns.lineplot(x="year",y="price",data=westdataset)
plt.xlabel("Years")
plt.ylabel("Prices")
plt.show()
```
